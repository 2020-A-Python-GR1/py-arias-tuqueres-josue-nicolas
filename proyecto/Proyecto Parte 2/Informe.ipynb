<h1 style="text-align: center">ACCIDENTES DE TRÁNSITO EN ECUADOR EN EL AÑO 2016</h1>


<p style="text-align:justify">Durante el año 2016, se registraron 30.269 siniestros por accidentes de tránsito. El promedio por mes fue de 2.522, siendo enero el mes con mayor número de accidentes con el 10,1% respecto del total. La provincia del Pichincha registró el mayor número de accidentes con el 35,6%, siguiéndole la provincia del Guayas con el 26,1% y en tercer lugar la provincia de Imbabura con el 5,1%. [1]</p>

<p style="text-align:justify">Las causas que provocaron el 77,7% de los accidentes fueron: impericia e imprudencia del conductor (51,9%), no respeta las señales de tránsito (13,4%) y exceso de velocidad (12,4%). [1]<p>

<p style="text-align:justify">Con respecto a la clase de accidentes, los choques constituyen el mayor número de accidentes con el 45,4%, siguiéndole los atropellos con el 16,8% y estrellamientos el 13,3%, los mismos que constituyen el 75,5% de accidentes ocurridos en el país. [1]<p>

<p style="text-align:justify">Como consecuencia se registra 23.425 víctimas en el 2016; de las cuales, el 91,6% fueron heridas, mientras que el 8,4% fallecieron. Las provincias que registran mayor número de víctimas son Guayas y Pichincha con el 33,4% y 26,5% respectivamente. [1]<p>

### Dataset

<p style="text-align: justify">Contiene información referente al número de accidentes de tránsito con respecto a: causas y clases del accidente, mes, provincia y número de fallecidos y lesionados; misma que fue proporcionada por la Agencia Nacional de Tránsito (ANT), respecto al año 2016. [1]</p>

### Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Carga de Dataset

```python
path_dataset = r"data\2016_Accidentes Tránsito_BDD.csv"
df = pd.read_csv(
    path_dataset)
print(df)
```

### Carga de Diccionario de datos

```python
path_diccionario = r"data\Diccionario Datos AT 2016.xlsx"
df_diccionario = pd.read_excel(path_diccionario)
```

#### Diccionario provincia

```python
diccionario_provincia = {int(item.split("=")[0]):item.split("=")[1] for item in df_diccionario.iloc[7,4].split('\n')}
#print(diccionario_provincia)
```

#### Diccionario canton

```python
diccionario_canton = {int(item.split("=")[0]):item.split("=")[1] for item in df_diccionario.iloc[8,4].split('\n')}
#print(diccionario_canton)
```

#### Diccionario mes


```python
diccionario_mes = {int(item.split("=")[0]):item.split("=")[1] for item in df_diccionario.iloc[9,4].split('\n')}
#print(diccionario_mes)
```

#### Diccionario dia de ocurrecia

```python
diccionario_dia_ocurrencia = {int(item.split("=")[0]):item.split("=")[1] for item in df_diccionario.iloc[10,4].split('\n')}
#print(diccionario_dia_ocurrencia)
```

#### Diccionario hora del suceso

```python
diccionario_hora_suceso = {int(item.split("=")[0]):item.split("=")[1] for item in df_diccionario.iloc[11,4].split('\n')}
#print(diccionario_hora_suceso)
```

#### Diccionario clase de accidente

```python
diccionario_clase = {int(item.split("=")[0]):item.split("=")[1] for item in df_diccionario.iloc[12,4].split('\n')}
#print(diccionario_clase)
```

#### Diccionario causa probable del accidente

```python
diccionario_causa_probable = {int(item.split("=")[0]):item.split("=")[1] for item in df_diccionario.iloc[13,4].split('\n')}
#print(diccionario_causa_probable)
```

#### Diccionario zona

```python
diccionario_zona = {int(item.split("=")[0]):item.split("=")[1] for item in df_diccionario.iloc[14,4].split('\n')}
#print(diccionario_zona)
```

### PREGUNTAS

<h2>1. Número de accidentes de tránsito por provincia</h2>

```python
accidentes_totales = int(len(df))
accidentes_por_provincia = df["PROVINCIA"].value_counts()
accidentes_por_provincia = accidentes_por_provincia.rename(diccionario_provincia)

def graficoBarrasHorizontal(nombres, datos, total, xlabel, titulo, yalign, fontsize):
    plt.rcdefaults()
    fig, ax = plt.subplots()

    name = list(nombres)
    data = list(datos)
    ax.barh(name, data, align='center')
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel(xlabel)
    ax.set_title(titulo)
    for i, v in enumerate(data):
        ax.text(v + 3, i + yalign, str("{0:0.2f}".format((v/total)*100)) + "%",  fontsize=fontsize)
        #ax.text(v + 3, i + yalign, str(v))
    plt.show()

graficoBarrasHorizontal(accidentes_por_provincia.index, 
                        accidentes_por_provincia, 
                        accidentes_totales,
                        'Numero de accidentes de tránsito',
                        "ACCIDENTES DE TRÁNSITO POR PROVINCIA",
                        0.30, 
                        9)
```

<h2>2. Los 10 cantones con más accidentes de tránsito</h2>

```python
accidentes_por_canton = df["CANTON"].value_counts()
accidentes_por_canton = accidentes_por_canton.rename(diccionario_canton)
graficoBarrasHorizontal(accidentes_por_canton.iloc[:10].index, 
                        accidentes_por_canton.iloc[:10], 
                        accidentes_totales,
                        'Numero de accidentes de tránsito',
                        "10 CANTONES CON MÁS ACCIDENTES DE TRÁNSITO", 
                        0.12, 
                        10)
```

<h2>3. Número de accidentes de tránsito por mes</h2>

```python
accidentes_por_mes = df["MES"].value_counts()
accidentes_por_mes = accidentes_por_mes.rename(diccionario_mes)
graficoBarrasHorizontal(accidentes_por_mes.index, 
                        accidentes_por_mes, 
                        accidentes_totales, 
                        'Numero de accidentes de tránsito',
                        "ACCIDENTES DE TRÁNSITO POR MES", 
                        0.15, 
                        10)
```

<h2>3. Número de accidentes de tránsito por día</h2>

```python
accidentes_por_dia = df["DIA"].value_counts()
accidentes_por_dia = accidentes_por_dia.rename(diccionario_dia_ocurrencia)
graficoBarrasHorizontal(accidentes_por_dia.index, 
                        accidentes_por_dia, 
                        accidentes_totales, 
                        'Numero de accidentes de tránsito',
                        "ACCIDENTES DE TRÁNSITO POR DIA", 
                        0.10, 
                        11)
```

<h2>4. Número de accidentes de tránsito por hora</h2>

```python
accidentes_por_hora = df["HORA"].value_counts()
accidentes_por_hora = accidentes_por_hora.rename(diccionario_hora_suceso)

graficoBarrasHorizontal(accidentes_por_hora.index, 
                        accidentes_por_hora, 
                        accidentes_totales,
                        'Numero de accidentes de tránsito',
                        "ACCIDENTES DE TRÁNSITO POR HORA", 
                        0.30, 
                        8)
```

<h2>5. Número de accidentes de tránsito por clase</h2>

```python
accidentes_por_clase= df["CLASE"].value_counts()
accidentes_por_clase = accidentes_por_clase.rename(diccionario_clase)

graficoBarrasHorizontal(accidentes_por_clase.index, 
                        accidentes_por_clase, 
                        accidentes_totales, 
                        'Numero de accidentes de tránsito',
                        "ACCIDENTES DE TRÁNSITO POR CLASE", 
                        0.10, 
                        11)
```

<h2>6. Número de accidentes de tránsito por causa probable</h2>

```python
accidentes_por_causa = df["CAUSA"].value_counts()
accidentes_por_causa = accidentes_por_causa.rename(diccionario_causa_probable)

graficoBarrasHorizontal(accidentes_por_causa.index, 
                        accidentes_por_causa, 
                        accidentes_totales, 
                        'Numero de accidentes de tránsito',
                        "ACCIDENTES DE TRÁNSITO POR CAUSA PROBABLE", 
                        0.15, 
                        11)
```

<h2>7. Número de accidentes de tránsito por zona</h2>

```python
accidentes_por_zona = df["ZONA"].value_counts()
accidentes_por_zona = accidentes_por_zona.rename(diccionario_zona)

graficoBarrasHorizontal(accidentes_por_zona.index, 
                        accidentes_por_zona, 
                        accidentes_totales, 
                        'Numero de accidentes de tránsito',
                        "ACCIDENTES DE TRÁNSITO POR ZONA", 
                        0.05, 
                        20)
```

<h2>8. Número de víctimas</h2>

```python
victima = df[["NUM_LESIONADO", "NUM_FALLECIDO"]]
NUM_LESIONADO = int(victima["NUM_LESIONADO"].sum())
NUM_FALLECIDO = int(victima["NUM_FALLECIDO"].sum())

indice_lesionados_fallecidos = ["Número de lesionados", "Número de fallecidos"]
data_lesionados_fallecidos = [NUM_LESIONADO, NUM_FALLECIDO]
victimas_totales = sum(data_lesionados_fallecidos)

graficoBarrasHorizontal(indice_lesionados_fallecidos, 
                        data_lesionados_fallecidos, 
                        victimas_totales, 
                        'Víctimas',
                        "NÚMERO DE VÍCTIMAS: " + str(victimas_totales), 
                        0.05, 
                        20)
```

<h2>9. Número de víctimas por provincia</h2>

```python
provincias =  list(diccionario_provincia.values())
lesionados_por_provincia = list(df[["PROVINCIA","NUM_LESIONADO"]].groupby("PROVINCIA").sum()["NUM_LESIONADO"])
fallecidos_por_provincia = list(df[["PROVINCIA","NUM_FALLECIDO"]].groupby("PROVINCIA").sum()["NUM_FALLECIDO"])
    
fig = plt.figure()
ax = fig.add_subplot(111)
ax.barh(provincias, lesionados_por_provincia, align='center', color='#00ff00',label='LESIONADOS')
ax.barh(provincias, fallecidos_por_provincia, align='center', left=lesionados_por_provincia, color='g',label='FALLECIDOS')
for i in range(len(lesionados_por_provincia)):
        ax.text(
            int(lesionados_por_provincia[i])+int(fallecidos_por_provincia[i]) + 10,  
            i - 0.3, 
            str( "{0:0.2f}".format(((int(lesionados_por_provincia[i])+int(fallecidos_por_provincia[i]))/victimas_totales)*100) ) + "%"
        )
ax.set_yticks(provincias)
ax.set_xlabel('VÍCTIMAS')
ax.set_title('NÚMERO DE VÍCTIMAS POR PROVINCIA')
ax.grid(True)
ax.legend()
plt.show()
```

<h2>10. Número de fallecidos por provincia</h2>

```python
fallecidos_por_provincia = df[["PROVINCIA","NUM_FALLECIDO"]].groupby("PROVINCIA").sum().rename(diccionario_provincia).sort_values(by=['NUM_FALLECIDO'],ascending=False)

indice_provincias = fallecidos_por_provincia.index
datos_fallecidos = fallecidos_por_provincia["NUM_FALLECIDO"]
total_fallecidos = sum(list(datos_fallecidos))

graficoBarrasHorizontal(indice_provincias, 
                        datos_fallecidos, 
                        total_fallecidos, 
                        'Fallecidos',
                        "NÚMERO DE FALLECIDOS POR PROVINCIA", 
                        0.30, 
                        9)
```

<h2>11. Número de lesionados por provincia</h2>

```python
lesionados_por_provincia = df[["PROVINCIA","NUM_LESIONADO"]].groupby("PROVINCIA").sum().rename(diccionario_provincia).sort_values(by=['NUM_LESIONADO'],ascending=False)

indice_provincias = lesionados_por_provincia.index
datos_lesionados = lesionados_por_provincia["NUM_LESIONADO"]
total_lesionados = sum(list(datos_lesionados))

graficoBarrasHorizontal(indice_provincias, 
                        datos_lesionados, 
                        total_lesionados, 
                        'Lesionados',
                        "NÚMERO DE LESIONADOS POR PROVINCIA", 
                        0.30, 
                        9)
```
