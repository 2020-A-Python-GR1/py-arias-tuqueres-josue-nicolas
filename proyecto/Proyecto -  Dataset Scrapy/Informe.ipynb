<h2>1. Las 10 marcas con más autos usados en venta</h2>

```python
def graficoBarrasHorizontal(nombres, datos, total, xlabel, titulo, xalign, yalign, fontsize, porcentaje):
    plt.rcdefaults()
    fig, ax = plt.subplots()

    name = list(nombres)
    data = list(datos)
    ax.barh(name, data, align='center')
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel(xlabel)
    ax.set_title(titulo)
    for i, v in enumerate(data):
        if porcentaje:
            ax.text(v + xalign, i + yalign, str("{0:0.2f}".format((v/total)*100)) + "%",  fontsize=fontsize) 
        else:
            ax.text(v + xalign, i + yalign, str(v), fontsize=fontsize)     
    ax.grid(True)
    plt.show()

autos_totales = int(len(df))

autos_por_marca = df["MARCA"].value_counts().head(10)

graficoBarrasHorizontal(autos_por_marca.index, 
                        autos_por_marca, 
                        autos_totales,
                        'Numero de autos usados en venta',
                        "LAS 10 MARCAS CON MÁS AUTOS USADOS EN VENTA\nAutos totales: " + str(autos_totales),
                        3,
                        0.10, 
                        9,
                        False)
```
