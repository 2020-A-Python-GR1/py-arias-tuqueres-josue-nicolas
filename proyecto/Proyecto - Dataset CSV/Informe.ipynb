<h1 style="text-align: center">ACCIDENTES DE TRÁNSITO EN ECUADOR EN EL AÑO 2016</h1>

### Dataset

<p style="text-align: justify">Contiene información referente al número de accidentes de tránsito con respecto a: causas y clases del accidente, mes, provincia y número de fallecidos y lesionados; misma que fue proporcionada por la Agencia Nacional de Tránsito (ANT), respecto al año 2016. [1]</p>

### Resumen


<p style="text-align:justify">Durante el año 2016, se registraron 30.269 siniestros por accidentes de tránsito. El promedio por mes fue de 2.522, siendo enero el mes con mayor número de accidentes con el 10,1% respecto del total. La provincia del Pichincha registró el mayor número de accidentes con el 35,6%, siguiéndole la provincia del Guayas con el 26,1% y en tercer lugar la provincia de Imbabura con el 5,1%. [1]</p>

<p style="text-align:justify">Las causas que provocaron el 77,7% de los accidentes fueron: impericia e imprudencia del conductor (51,9%), no respeta las señales de tránsito (13,4%) y exceso de velocidad (12,4%). [1]<p>

<p style="text-align:justify">Con respecto a la clase de accidentes, los choques constituyen el mayor número de accidentes con el 45,4%, siguiéndole los atropellos con el 16,8% y estrellamientos el 13,3%, los mismos que constituyen el 75,5% de accidentes ocurridos en el país. [1]<p>

<p style="text-align:justify">Como consecuencia se registra 23.425 víctimas en el 2016; de las cuales, el 91,6% fueron heridas, mientras que el 8,4% fallecieron. Las provincias que registran mayor número de víctimas son Guayas y Pichincha con el 33,4% y 26,5% respectivamente. [1]<p>

<h2>1. Número de accidentes de tránsito, según provincias</h2>

```python
accidentes_totales = int(len(df))
accidentes_por_provincia = df["PROVINCIA"].value_counts().rename(diccionario_provincia)

def graficoBarrasHorizontal(nombres, datos, total, xlabel, titulo, yalign, fontsize, porcentaje):
    plt.rcdefaults()
    fig, ax = plt.subplots()

    name = list(nombres)
    data = list(datos)
    ax.barh(name, data, align='center')
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel(xlabel)
    ax.set_title(titulo)
    for i, v in enumerate(data):
        if porcentaje:
            ax.text(v + 3, i + yalign, str("{0:0.2f}".format((v/total)*100)) + "%",  fontsize=fontsize) 
        else:
            ax.text(v + 3, i + yalign, str(v), fontsize=fontsize)     
    ax.grid(True)
    plt.show()

graficoBarrasHorizontal(accidentes_por_provincia.index, 
                        accidentes_por_provincia, 
                        accidentes_totales,
                        'Numero de accidentes de tránsito',
                        "ACCIDENTES DE TRÁNSITO POR PROVINCIA\nAccidentes totales: " + str(accidentes_totales),
                        0.30, 
                        9,
                        True)
```

```python
victima = df[["NUM_LESIONADO", "NUM_FALLECIDO"]]
NUM_LESIONADO = int(victima["NUM_LESIONADO"].sum())
NUM_FALLECIDO = int(victima["NUM_FALLECIDO"].sum())

indice_lesionados_fallecidos = ["Número de lesionados", "Número de fallecidos"]
data_lesionados_fallecidos = [NUM_LESIONADO, NUM_FALLECIDO]
victimas_totales = sum(data_lesionados_fallecidos)

graficoBarrasHorizontal(indice_lesionados_fallecidos, 
                        data_lesionados_fallecidos, 
                        victimas_totales, 
                        'Víctimas',
                        "NÚMERO DE VÍCTIMAS: " + str(victimas_totales), 
                        0.05, 
                        20,
                        False)
```

```python
provincias =  list(diccionario_provincia.values())
lesionados_por_provincia = list(df[["PROVINCIA","NUM_LESIONADO"]].groupby("PROVINCIA").sum()["NUM_LESIONADO"])
fallecidos_por_provincia = list(df[["PROVINCIA","NUM_FALLECIDO"]].groupby("PROVINCIA").sum()["NUM_FALLECIDO"])
    
fig = plt.figure()
ax = fig.add_subplot(111)
ax.barh(provincias, lesionados_por_provincia, align='center', color='#00ff00',label='LESIONADOS')
ax.barh(provincias, fallecidos_por_provincia, align='center', left=lesionados_por_provincia, color='g',label='FALLECIDOS')
for i in range(len(lesionados_por_provincia)):
        ax.text(
            int(lesionados_por_provincia[i])+int(fallecidos_por_provincia[i]) + 10,  
            i - 0.3, 
            str( "{0:0.2f}".format(((int(lesionados_por_provincia[i])+int(fallecidos_por_provincia[i]))/victimas_totales)*100) ) + "%"
        )
ax.set_yticks(provincias)
ax.set_xlabel('VÍCTIMAS')
ax.set_title('NÚMERO DE VÍCTIMAS POR PROVINCIA\nVíctimas totales: ' + str(victimas_totales))
ax.grid(True)
ax.legend()
plt.show()
```
